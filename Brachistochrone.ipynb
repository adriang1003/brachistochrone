```python
# File: Brachistochrone.ipynb
# Author: Adrian Garcia
# Purpose: Solves the Brachistochrone problem for a vertically varying point A
#          and a horizontally fixed point B. NOTE: This problem was solved
#          analytically; thus y_A, c_1, and theta_B were hard coded.
import matplotlib.pyplot as plt
import numpy as np
import math

plt.rcParams["figure.figsize"] = (10, 8)
plt.rcParams.update({'font.size': 12})
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXgeneral'

# Variable declaration
y_A = [0.1, 0.2, 0.3, 0.4, 1, 7, 8, 9, 10]
x_B = 1
g = 9.8
c_1 = [0.331239, 0.357932, 0.397142, 0.449746, 1.14583, 156.661, 232.37, 329.413, 450.456]
theta_B = [-5.11977, -4.59459, -4.17628, -3.81967, -2.41201, -0.425978, -0.373258, -0.332108, -0.299105]
theta = np.linspace(-2 * math.pi, 0, 1000)

# Travel time
T = []
for i in range(len(c_1)):
    T.append(-math.sqrt(c_1[i] / (2 * g)) * theta_B[i])
for j in range(len(y_A)):
    print('(0, {}) T = {:1.3f}'.format(y_A[j], T[j]))

# Brachistochrone curve
x = []
y = []
for k in theta:
    x.append(-(c_1[4]/2)*(k - math.sin(k)))
    y.append(y_A[4] - (c_1[4]/2)*(1 - math.cos(k)))
```

```python
# Plot 1
fig = plt.figure()
plt.plot(x, y, 'green', label = 'optimal curve $T = 0.583$ sec')

# Plot 1 config
plt.xlim(0, x_B)
plt.ylim(0, y_A[4])
plt.title('Optimal Curve: $(0,1)$ to $(1,0)$', fontsize = 24)
plt.xlabel('$x$', fontsize = 18)
plt.ylabel('$y$', fontsize = 18, rotation = 0)
plt.legend()
plt.grid(True)
plt.show()
fig.savefig('fig_1', dpi = 600)
```

```python
def line(x):
    y = []
    for i in x:
        y.append(-i + 1)
    return y

def parabola(x):
    y = []
    for i in x:
        y.append((i - 1)**2)
    return y

def circle(x):
    y = []
    for i in x:
        y.append(-(1 - (i - 1)**2)**(1/2) + 1)
    return y

x_1 = np.linspace(0, 1, 1000)
# Plot 2
fig = plt.figure()
plt.plot(x_1, line(x_1), label = 'straight line $T = 0.639$ sec')
plt.plot(x_1, parabola(x_1), label = 'parabola $T = 0.595$ sec')
plt.plot(x, y, 'green', label = 'optimal curve $T = 0.583$ sec')
plt.plot(x_1, circle(x_1), 'purple', label = 'circle $T = 0.592$ sec')

# Plot 2 config
plt.xlim(0, x_B)
plt.ylim(0, y_A[4])
plt.title('Optimal Curve Comparison: $(0,1)$ to $(1,0)$', fontsize = 24)
plt.xlabel('$x$', fontsize = 18)
plt.ylabel('$y$', fontsize = 18, rotation = 0)
plt.legend()
plt.grid(True)
plt.show()
fig.savefig('fig_2', dpi = 600)
```
